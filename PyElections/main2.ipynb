```python
#Scope of work - write a script that will decide the two candidates with the highest number
#of votes who will advance to the runoff election.

#The workflow is described as follows:

#1. Open the CSV
import os
import csv
#set path for file
csvpath = os.getcwd()+"\houston_election_data.csv"

#Create a dictionary from the CSV file to be printed at the end of this step
Dict_Election = {}
    
with open(csvpath, encoding = 'utf-8') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

# Read the header row first    
    csv_header = next(csvreader)
    
#Read each row after the header
    next(csvreader)    
    for row in csvreader:
        Dict_Election[row[2]] = [row[0], row[1]]

#2. Determine the total number of votes cast
#with set we don't repeat the number of voters

voters = set(Dict_Election.keys())
voters_total = len(voters)

#Print the first part of the Results Summary Table

print("")
print("Houston Mayoral Election Results")
print("-" * 42)
print("Total Cast Votes: {}".format(voters_total))
print("-" * 42)

#3. A complete list of candidates who received votes
#Create a second summarized dictionary (empty at first)

dict_total_votes = {} 
for key, value in Dict_Election.items():
    if Dict_Election[key][0] not in dict_total_votes:
        dict_total_votes[Dict_Election[key][0]] = 1
    else:
        dict_total_votes[Dict_Election[key][0]] += 1

#4. The percentage of votes and the total number each candidate won 

for key, value in dict_total_votes.items():
    average_percentage = (value / voters_total) * 100
    #Print the middle of the Results Summary Table
    print("{}: {:.2f}% ({})".format(key, average_percentage, value))

print("-" * 42)

#5. Print the names of the two candidates who will advance to the runoff election.

maximum_votes = max(dict_total_votes.values())
#initializing the 2nd max
second_maximum_votes = 0
#Defining the loop to get the 2nd advancing candidate
for key, value in dict_total_votes.items():
    if value > second_maximum_votes and value < maximum_votes:
        second_maximum_votes = value

#Defining the values to be printed in the final part of the Results Summary Table
for key, value in dict_total_votes.items():
    if maximum_votes == value:
        first_advancing = key
    elif second_maximum_votes == value:
        second_advancing = key
        
print("1st Advancing Candidate: {}".format(first_advancing))
print("2nd Advancing Candidate: {}".format(second_advancing))
print("-" * 42)
```
